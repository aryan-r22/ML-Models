```
import tensorflow as tf
```

```
import numpy as np
```

```
import matplotlib.pyplot as plt
```

```
features=np.array([0,3,10,26,45,70])
```

```
label=np.array([32,37.4,50,78.8,113,158])
```

```
l0=tf.keras.layers.Dense(units=4,input_shape=(1,))
```

```
l1=tf.keras.layers.Dense(units=5)
```

```
l2=tf.keras.layers.Dense(units=1)
```

```
model=tf.keras.Sequential([l0,l1,l2])
```

```
model.compile(loss='mean_squared_error',optimizer=tf.keras.optimizers.Adam(0.05))
```

```
history=model.fit(features,label,epochs=300)
```

```
plt.plot(history.history['loss'])
plt.xlabel="epochs"
plt.ylabel="error"
```

```
model.predict([9])
```
